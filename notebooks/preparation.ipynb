# Library and Configuration

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import kagglehub
import os
from pathlib import Path

from rapidfuzz import process

path = Path(kagglehub.dataset_download("olistbr/brazilian-ecommerce"))
print(os.listdir(path))

print("Library loaded!")
```

# Load Data

```python
df_customers      = pd.read_csv(path / "olist_customers_dataset.csv")
df_geolocation    = pd.read_csv(path / "olist_geolocation_dataset.csv")
df_orders         = pd.read_csv(path / "olist_orders_dataset.csv")
df_order_items    = pd.read_csv(path / "olist_order_items_dataset.csv")
df_order_payments = pd.read_csv(path / "olist_order_payments_dataset.csv")
df_order_reviews  = pd.read_csv(path / "olist_order_reviews_dataset.csv")
df_products       = pd.read_csv(path / "olist_products_dataset.csv")
df_sellers        = pd.read_csv(path / "olist_sellers_dataset.csv")
df_category_translation = pd.read_csv(
    path / "product_category_name_translation.csv"
)
print("Data Loaded!")
```

```python
df_orders["order_status"].value_counts()
```

order status valid

```python
datetime_cols = df_orders.filter(like='date').columns.union(
    df_orders.filter(like='timestamp').columns.union(
    df_orders.filter(like='approved').columns)
)

before_na = df_orders[datetime_cols].isna().sum()

df_orders[datetime_cols] = df_orders[datetime_cols].apply(
    pd.to_datetime,
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)

after_na = df_orders[datetime_cols].isna().sum()

pd.DataFrame({
    'before_na': before_na,
    'after_na': after_na,
    'new_invalid': after_na - before_na
})
```

format order date dan timestamp konsisten

```python
df_orders["order_id"].nunique() == len(df_orders)
```

valid bahwa order id bernilai unik

```python
date_order_checks = {
    'approved_before_purchase':
        (df_orders['order_approved_at'] < df_orders['order_purchase_timestamp']).sum(),

    'carrier_before_approved':
        (df_orders['order_delivered_carrier_date'] < df_orders['order_approved_at']).sum(),

    'delivered_before_carrier':
        (df_orders['order_delivered_customer_date'] < df_orders['order_delivered_carrier_date']).sum(),

    'estimated_before_purchase':
        (df_orders['order_estimated_delivery_date'] < df_orders['order_purchase_timestamp']).sum(),

    'delivered_after_estimated':
        (df_orders['order_delivered_customer_date'] > df_orders['order_estimated_delivery_date']).sum(),
}

pd.Series(date_order_checks)
```

```python
(
    df_orders['order_approved_at']
    - df_orders['order_delivered_carrier_date']
).loc[
    df_orders['order_delivered_carrier_date']
    < df_orders['order_approved_at']
].describe()
```

terdapat lebih dari 1000 baris yang memiliki urutan waktu dari tabel yang tidak sesuai, akan dijadikan validasi data

```python
print(df_order_payments["order_id"].nunique())
print(df_order_payments["payment_sequential"].value_counts().sort_values())
```

```python
check = (
    df_order_payments
    .groupby('order_id')['payment_sequential']
    .apply(lambda x: sorted(x.tolist()) == list(range(1, x.max()+1)))
)

check.value_counts()
```

```python
invalid = (
    df_order_payments
    .groupby('order_id')['payment_sequential']
    .apply(lambda x: sorted(x.tolist()) != list(range(1, x.max()+1)))
)

invalid_orders = invalid[invalid].index

df_order_payments[
    df_order_payments['order_id'].isin(invalid_orders)
].sort_values(['order_id', 'payment_sequential'])
```

```python
multi_cc = (
    df_order_payments
    .groupby('order_id')
    .filter(
        lambda x:
        (x['payment_type'] == 'credit_card').sum() > 1
    )
)

multi_cc.sort_values(['order_id', 'payment_sequential']).head()
```
